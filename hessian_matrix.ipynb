```python
import cv2
import numpy as np
from scipy.ndimage import gaussian_laplace
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/Screenshot 2023-09-28 at 17.39.13.png',cv2.IMREAD_GRAYSCALE)
```

```python
HESSUAN = gaussian_laplace(image,sigma=1)
```

```python
threshold =0.1
```

```python
blobs = HESSUAN > threshold
```

```python
image_with_blobs = cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
image_with_blobs[blobs] = [0,0,255]
```

```python
cv2_imshow(image_with_blobs)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
#LOAD AN IMAGE
image = cv2.imread('/content/Screenshot 2023-09-28 at 17.39.13.png',cv2.IMREAD_COLOR)
```

```python
#convert the image to grayscale
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
#create a Sift detector object
sift = cv2.SIFT_create()
```

```python
#detect keyponts ans compute Sift
keypoints,descriptor = sift.detectAndCompute(gray_image,None)
```

```python
image_with_keypoints = cv2.drawKeypoints(image,keypoints,outImage=None)
```

```python
cv2_imshow(image_with_keypoints)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
