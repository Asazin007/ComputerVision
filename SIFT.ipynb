SIFT

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
#load an image
image = cv2.imread('img1.jpg',cv2.IMREAD_COLOR)

#convert this to grayscale
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

#create a sift detector
sift= cv2.SIFT_create()
#detect keyponts ans compute Sift
keypoints,descriptor = sift.detectAndCompute(gray_image,None)
image_with_keypoints = cv2.drawKeypoints(image,keypoints,outImage=None)
cv2_imshow(image_with_keypoints)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
